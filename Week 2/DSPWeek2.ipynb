```python
%matplotlib notebook
# Above is a 'magic' line and will provide interactive plots
import numpy as np
from scipy.io import wavfile
from IPython.display import Audio
import matplotlib.pyplot as plt
import math
```

```python
# Read the wav file
Fs, data = wavfile.read('week2_unknown_guitar_string.wav')
```

```python
print("Sampling rate: {}".format(Fs))

# Show audio player
Audio(data, rate=Fs)
```

```python
T = 1.0/Fs
spec = np.fft.fft(data)
N = len(spec)
spec = spec/N

xf = np.fft.fftfreq(N, T)
plt.figure()
plt.plot(xf, np.abs(spec))
plt.title('frequency spectrum of the guitar string')
plt.xlim([0,1000])
plt.xlabel('frequency(Hz)')
plt.ylabel('amplitude')
plt.show()
#it shows B string (approx. 246 Hz)
```
